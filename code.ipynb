```python
import json
import random
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from sklearn.preprocessing import LabelEncoder
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import nltk
import pickle
import os
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
# Load the dataset
with open('intents.json', 'r') as file:
    data = json.load(file)
```

```python
# Preprocess the data
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

words = []
classes = []
documents = []
```

```python
for intent in data['intents']:
    for pattern in intent['patterns']:
        # Tokenize and preprocess the pattern
        tokens = word_tokenize(pattern)
        tokens = [stemmer.stem(word.lower()) for word in tokens if word not in stop_words]
        words.extend(tokens)
        documents.append((tokens, intent['tag']))

    if intent['tag'] not in classes:
        classes.append(intent['tag'])

words = sorted(set(words))
classes = sorted(set(classes))
```

```python
# Save words and classes for later use
with open('words.pkl', 'wb') as file:
    pickle.dump(words, file)
with open('classes.pkl', 'wb') as file:
    pickle.dump(classes, file)
```

```python
# Create training data
training = []
output_empty = [0] * len(classes)

for tokens, tag in documents:
    bag = [1 if word in tokens else 0 for word in words]
    output_row = output_empty[:]
    output_row[classes.index(tag)] = 1
    training.append((bag, output_row))

random.shuffle(training)
training = np.array(training, dtype=object)

train_x = np.array(list(training[:, 0]))
train_y = np.array(list(training[:, 1]))
```

```python
# Build the model
model = Sequential([
    Dense(128, input_shape=(len(train_x[0]),), activation='relu'),
    Dropout(0.5),
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(len(train_y[0]), activation='softmax')
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(train_x, train_y, epochs=200, batch_size=8, verbose=1)
```

```python
# Save the model
model.save('my_model.keras')
```

```python
# Chat function
def clean_up_sentence(sentence):
    sentence_words = word_tokenize(sentence)
    sentence_words = [stemmer.stem(word.lower()) for word in sentence_words if word not in stop_words]
    return sentence_words

def bow(sentence, words):
    sentence_words = clean_up_sentence(sentence)
    bag = [1 if word in sentence_words else 0 for word in words]
    return np.array(bag)

def chat():
    print("Start talking with the chatbot (type 'quit' to stop):")
    while True:
        message = input("You: ")
        if message.lower() == 'quit':
            break

        bag_of_words = bow(message, words)
        results = model.predict(np.array([bag_of_words]))[0]
        results_index = np.argmax(results)
        tag = classes[results_index]

        if results[results_index] > 0.7:
            for intent in data['intents']:
                if intent['tag'] == tag:
                    print(random.choice(intent['responses']))
        else:
            print("I didn't understand that. Can you please rephrase?")
```

```python
chat()
```
